```python
import pandas as pd
import re

df = pd.read_csv('cleaned_data.csv')
```

```python
df = df[df['transcript'].notnull()]
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyzer_scores(s):
    scores = analyser.polarity_scores(s)
    return scores['compound']

df['sentiment'] = df['transcript'].map(sentiment_analyzer_scores)
```

```python
import matplotlib.pyplot as plt
plt.axis('off')
plt.scatter('views', 'sentiment', data = df, color = 'red')
```

```python
df_p = df[df['sentiment'] > 0]
df_n = df[df['sentiment'] < 0]
p = df['sentiment'][df['sentiment'] > 0].count()
n = df['sentiment'][df['sentiment'] < 0].count()
df_credit = pd.Series({'Positive':p,'Negative':n})
df_credit.plot(kind = 'pie')
```
