```python
import pandas as pd
import re

df = pd.read_csv('Transcripts.csv')
```

```python
def speaker(s):

    speaker = re.findall('Subtitles(.*) \|',s)[0].strip('  ')

    return speaker

def views(s):

    views = re.findall('([0-9,]*) views',s)[0]
    
    return views

def duration(s):

    duration = re.findall('([0-9]*:[0-9]*)',s)[1]

    return duration

def subtitle(s):

    subtitle = re.findall('Subtitles in ([0-9]*) languages',s)[0]
    
    return subtitle

df['speaker'] = df['Field1'].apply(speaker)
df['views'] =  df['Field1'].apply(views)
df['duration'] = df['Field1'].apply(duration)
df['subtitle']  =  df['Field1'].apply(subtitle)
df.head(9)
```

```python
def content(s):
    try:
        new = re.findall('0:00(.*)',s.strip('0:00'))[0]
        new1 = re.sub(r'([0-9]*:[0-9]*)', '',new)
        new2 = re.sub(r'\xa0',' ',new1)
        new3 = re.sub(r'What inspires you?(.*)','',new2)
        new4 = re.sub(r'\(.*?\)','',new3).replace('\\','')
        new4 = re.sub(r'\[.*?\]','',new4).replace('\\','')
        return new4
    
    except:
        return None
        
df['transcript'] = df['Field2'].apply(content)
```

```python
df.to_csv('cleaned_data.csv')
```
